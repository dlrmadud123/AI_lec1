### 3.3 변수, 연산, 자료형
####3.3.1 변수와 연산
프로그램에는 데이터를 보관하는 저장소가 필요한데 데이터를 보관하는 곳을 변수라고 합니다. C언어와 같은 다른 프로그램과 달리 파이썬은 변수에 담는 데이터의 형태 (정수, 부동소수, 문자 등)에 따라 변수 선언을 다르게 할 필요가 없습니다.

````python
value = 2021  # 숫자
# 이것은 주석입니다.  
name = 'Hi'   # 문자
print(value)
print(name) 
````

가장 기본적인 연산은 사칙연산입니다. (+,-,*,/)
그 외에 자주 사용되는 연산 기호는 다음과 같습니다. 
- ** : 제곱 (ex) 2 ** 4 -> 16
- // : 정수로 나누었을 때의 몫
- % : 정수로 나누었을 때의 나머지
아래 결과를 확인해보길 바랍니다.

````python
a = 12
b = 5

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a ** b)
print(a % b)
````

#### 3.3.2 자료형

숫자의 종류에는 정수형과 실수형이 있습니다. 파이썬에서는 상관없이 변수에 넣어주면 됩니다.
1.5e4 같은 경우에는 컴퓨터의 지수 표현 방식으로 1.5*10^4 를 나타냅니다. 



````python
a = -1.124
b = 1.5e4
c = 1.5e-4
print(a,b,c)
````

문자열 인덱싱과 슬라이싱

인덱싱(Indexing)이란 어떤 것을 뽑아낸다는 의미이고, 슬라이싱(Slicing)은 잘라낸다는 의미입니다.

문자열은 첫 번째 자리부터 0,1,2,3,4 ... 번째로 번호를 매길 수 있습니다. 공백을 포함하며 a[0]는 a라는 문자열의 0번째 문자를 나타냅니다. 

아래 예제에서 0번째 문자는 m입니다. a[-3]은 뒤에서 3번째 문자를 가리키며 'T'입니다. 

문자열 슬라이싱은 여러 개의 문자를 뽑아내고 싶을 때 사용하면 됩니다. a[3:7]은 3,4,5,6 번째 문자를 나타냅니다. 앞 번호를 생략하면 처음부터, 뒷 번호를 생략하면 끝까지 뽑아냅니다. 

````python
# 문자열 인덱싱과 슬라이싱
a = 'my name is Tan'
print(a[4])
print(a[-3])
print(a[3:7])
print(a[4:])
print(a[:5])
````

**리스트 (List)**

리스트를 사용하면 여러 모음을 하나의 변수로 저장할 수 있습니다. 

리스트명 = [요소1, 요소2, 요소3, 요소4, ...] 

이렇게 만들어주면 하나의 리스트에 여러 요소들이 담기게 됩니다. 앞에서 배운 숫자, 문자열이 요소가 될 수 있고 리스트도 리스트의 요소가 될 수 있습니다. 


````python
# 빈 리스트 
a = []
# 문자열 리스트
b = ['python', 'is', 'fun']

# 숫자와 문자열 리스트
c = [1, 2, 3, 5, 8, 'fibonnaci']
# 리스트 c를 요소로 갖는 리스트
d = ['list', c]
print(a, b, c)
print(d)
````

앞서 문자열과 같이 리스트도 인덱싱을 할 수 있습니다. A=[1, 3, 5, 7, ['odd', 'numers']]인 리스트 A가 있으면, A[0]는 0번째 요소인 1을 나타냅니다.
리스트 요소 중 A[4]는 ['odd', 'numbers']가 되는데 여기서 첫 번째 요소를 가져오고 싶으면 A[4][0]으로 표현하면 됩니다.

````python
a = [1, 3, 5, 7, ['odd','numbers']]
print(a[0])
print(a[4])
print(a[4][1])
````

리스트에는 리스트를 편하게 가공하기 위해 다양한 내장 함수를 제공합니다. 

````python
a = [1, 3, 5, 7, 100, 100.5, 941.2, 18.51]

# 리스트의 길이 구하기
print(len(a))
# 리스트의 min, max
print(min(a))
print(max(a))
# 리스트의 합
print(sum(a))
# 오름차순 정렬
print(sorted(a))

b = [1, 2, 3, 4]
name = ['Sugar', 'JHO', 'Mina', 'Yeji', 'Mina']
# 리스트 요소 삭제
del b[1]
print(b)
# 리스트 요소 추가
b.append(5)
print(b)
# 리스트 요소 삽입
b.insert(1,2.5)
print(b)
# 리스트에 여러 요소 삽입
b.extend([6, 7, 8, 9])
print(b)
# 리스트 순서 뒤집기
name.reverse()
print(name)
# 요소가 저장된 인덱스 반환
print(name.index('Mina'))
# 요소의 개수 반환 
print(name.count('Mina'))
````

**튜플 (Tuple)**

튜플은 리스트와 거의 비슷하지만 다음 두 가지가 다릅니다.

1) 리스트는 []를 사용하지만 튜플은 ()를 사용한다. 

2) 리스트는 요소들을 생성, 삭제, 수정 할 수 있지만 튜플은 요소 값들을 바꿀 수 없다. 

따라서, 값이 항상 변하지 않는 것을 원할 때 튜플에 값을 저장하면 됩니다. 

인덱싱과 슬라이싱은 리스트와 동일하게 할 수 있습니다. 


````python
a = ()
b = (1,)
c = (1, 2, 3)
d = 1, 2, 3
e = ('a', 'b', ('ab', 'cd'))
print(e[1:])
print(len(e))
````

**딕셔너리 (Dictionary)**

딕셔너리는 Key와 Value로 이루어진 자료형입니다. 예를 들어 사람은 '이름', '나이','취미' 라는 Key 값을 갖고 이에 대응되는 '이지은', '15', '노래' 값을 가질 수 있습니다. 이런 식으로 대응 관계가 포함된 배열을 저장할 때 딕셔너리를 사용하면 됩니다. 

{Key1:Value1, Key2:Value2, Key3:Value3, ...}


````python
dic = {'name':'IU', 'age':15, 'hobby':'sing'}
dic
````

여기에 추가적으로 'height'라는 Key와 Value 쌍을 추가하고 싶다면

````python
dic['height'] = 160
dic
````

딕셔너리는 이렇게 어떤 쌍을 이루는 데이터를 저장할 때 매우 유용합니다. 예를 들어, 네 사람의 점수를 저장하고 싶다면

score = {'IU':90, 'sugar':80, 'sana':75, 'winter':85}

이렇게 저장할 수 있습니다. 특정 key의 value를 알고 싶다면 score[key] 처럼 사용하면 됩니다. 

````python
score = {'IU':90, 'sugar':80, 'sana':75, 'winter':85}
print(score['IU'])
print(score['winter'])
````

딕셔너리 키:요소 추가
딕셔너리 요소 제거 


````python
# key 뽑아내기
score.keys()
print(score.keys())
# value 뽑아내기
score.values()
print(score.values())
# key로 value 뽑아내기
a = score.get('rm')
print(a)
````

**넘파이(Numpy)**

Numpy는 행렬을 이용하여 수치 데이터를 다룰 때 매우 자주 사용되는 파이썬 패키지입니다. 


````python
import numpy as np
````

Numpy의 주요 모듈
1. np.array()  : 리스트, 튜플, 배열로 부터 ndarray를 생성
2. np.asarray(): 기존의 array로 부터 ndarray를 생성
3. np.arange() : range와 비슷
4. np.linspace(start, end, num) : [start, end] 균일한 간격으로 num개 생성
5. np.logspace(start, end, num) : [start, end] log scale 간격으로 num개 생성

````python
a = np.array([1, 2, 3, 4, 5]) #리스트를 가지고 1차원 배열 생성
print(type(a))
print(a)
````

````python
b = np.array([[10, 20, 30], [ 60, 70, 80]]) 
print(b)
````

````python
print(b.ndim)   # 몇 차원 행렬?
print(b.shape)  # 어떤 모양의 행렬?
````
